```python
import anndata as ad
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import scanpy as sc
import seaborn as sns

import os
import sys
main_path = os.getcwd()
```

```python
adata_path = os.path.join(main_path, 'data/ssl_adata/pdl1-pos/adata-train.hdf5')
adata_train = ad.read_h5ad(adata_path)
```

```python
sns.set_theme(style="white")
sns.set_context("paper", font_scale=3.0)
fig, ax = plt.subplots(figsize=(10,10))
sc.pl.umap(adata_train, color='leiden', show=False, size=30, legend_loc='on data', palette='tab20', ax=ax)
ax.set_title("")
ax.axis('off')
plt.tight_layout()
plt.show()
```
